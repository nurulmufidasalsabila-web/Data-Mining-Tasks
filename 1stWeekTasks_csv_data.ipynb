```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/drive/MyDrive/student.csv')
```

```python
data
```

```python
Grade_counts = data['Grade'].value_counts().sort_index()
```

```python
plt.figure()
plt.pie(
    Grade_counts,
    labels=Grade_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Distribusi Grade Mahasiswa')
plt.axis('equal')
plt.show()
```

```python
plt.figure()
plt.bar(Grade_counts.index, Grade_counts.values)
plt.xlabel('Grade')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Distribusi Grade Mahasiswa')

plt.show()
```

```python
grade_hours = data.groupby('Grade')['Weekly_Study_Hours'].sum()
```

```python
plt.figure()
plt.pie(
    grade_hours,
    labels=grade_hours.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Distribusi Weekly Study Hours berdasarkan Grade')
plt.axis('equal')
plt.show()
```

```python
age_by_scholarship = data.groupby('Scholarship')['Student_Age'].mean()
```

```python
plt.figure()
plt.bar(
    age_by_scholarship.index,
    age_by_scholarship.values
)
plt.title('Rata-rata Usia Mahasiswa berdasarkan Status Scholarship')
plt.xlabel('Scholarship')
plt.ylabel('Rata-rata Student Age')
plt.show()
```
